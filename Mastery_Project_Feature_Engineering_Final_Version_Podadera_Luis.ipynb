# **Features from Sessions**

```python
# Amount of trips per user for the first user. Contained are both cancelled and non-cancelled
amount_of_trips = sessions[['trip_id', 'user_id']].groupby('user_id').nunique()
amount_of_trips.columns = ['total_trips']
```

```python
amount_of_trips
```

```python
# Amount of cancelled trips per user (need to fillna(0) later when merging)
canceled_trips_count = sessions[sessions['cancellation'] == 1][['trip_id','user_id']].groupby('user_id').nunique()
canceled_trips_count.columns = ['canceled_trips']
```

```python
# Trip id of canceled trips
canceled_trip_ids = sessions[sessions['cancellation'] == 1]['trip_id'].unique()
```

```python
canceled_trips_count
```

```python
# Sessions of trips that actually happened or trips that were booked and cancelled
valid_sessions = sessions[~sessions['trip_id'].isin(canceled_trip_ids)]
#valid meaning no cancelled trips
```

```python
valid_sessions
```

```python
#check duplicates
valid_sessions.duplicated().sum()
```

```python
# Df to relate user_id with trip_id which will be used later with hotels and flights
user_trip_id = valid_sessions[['user_id', 'trip_id']].drop_duplicates()
user_trip_id
#seeing lots of NaN trip_id
```

```python
# Use only non-cancelled, completed bookings for sessions features
valid_bookings = valid_sessions[valid_sessions['trip_id'].notna()]
valid_bookings
```

```python
browsing_sessions = valid_sessions[valid_sessions['trip_id'].isna()]
browsing_sessions #sessions where no booking was made, still contains important user information
```

```python
browsing_users = browsing_sessions['user_id'].nunique()
print(f"Users who browsed: {browsing_users}") #almost everyone had at least one only browsing session to check out the website
```

```python
# Users with at least one trip
bookers = valid_bookings['user_id'].unique()

# Users who only browsed
non_bookers = set(browsing_sessions['user_id']) - set(bookers)

print(f"Users who only browsed (never booked): {len(non_bookers)}")
#540 lurkers
```

```python
# Step 1: Create the DataFrame from bookers
has_booked = pd.DataFrame({'user_id': bookers, 'has_booked': 1})

# Step 2: Merge with all users to ensure all user_ids are included
all_users = pd.DataFrame({'user_id': sessions['user_id'].unique()})
has_booked = all_users.merge(has_booked, on='user_id', how='left')

# Step 3: Fill non-bookers with 0 and set user_id as index
has_booked['has_booked'] = has_booked['has_booked'].fillna(0).astype(int)
has_booked = has_booked.set_index('user_id')
#this flags users as having atleast booked one trip or none at all.
```

```python
# Count number of browsing-only sessions per user
browsing_counts = browsing_sessions.groupby('user_id').size().rename('browsing_sessions')

# Count number of real bookings per user
booking_counts = valid_bookings.groupby('user_id').size().rename('booking_sessions')

# Combine them
browsing_behavior = pd.concat([browsing_counts, booking_counts], axis=1)

# Flag: User has both browsing and bookings
browsing_behavior['browsed_and_booked'] = (
    (browsing_behavior['browsing_sessions'] > 0) &
    (browsing_behavior['booking_sessions'] > 0)
).astype(int)

browsing_behavior = browsing_behavior.fillna(0).astype(int)
```

```python
only_browsed = has_booked.copy()
only_browsed['only_browsed'] = (only_browsed['has_booked'] == 0).astype(int)
only_browsed = only_browsed[['only_browsed']]
```

```python
browsing_behavior
```

```python
# 1. Combine the counts (browsing and booking)
session_counts = pd.concat([browsing_counts, booking_counts], axis=1)

# 2. Fill NaNs first to ensure arithmetic works
session_counts[['browsing_sessions', 'booking_sessions']] = session_counts[[
    'browsing_sessions', 'booking_sessions'
]].fillna(0)

# 3. Now safe to compute total sessions
session_counts['total_sessions'] = (
    session_counts['browsing_sessions'] + session_counts['booking_sessions']
)

# 4. Now cast to int (after safe totals)
session_counts[['browsing_sessions', 'booking_sessions', 'total_sessions']] = session_counts[[
    'browsing_sessions', 'booking_sessions', 'total_sessions'
]].astype(int)

# 5. Now calculate the booking ratio safely
session_counts['booking_ratio'] = (
    session_counts['booking_sessions'] / session_counts['total_sessions']
).replace([np.inf, -np.inf], 0).fillna(0).round(3)
```

```python
session_counts
```

```python
# Should be equal:
print("Sum of total_sessions:", session_counts['total_sessions'].sum())
print("Rows in valid_sessions:", valid_sessions.shape[0])
```

```python
# Make sure both are Series or DataFrames indexed by user_id
# If they are not already, set the index:
# amount_of_trips = amount_of_trips.set_index('user_id')
# canceled_trips_count = canceled_trips_count.set_index('user_id')

# Join them into a single DataFrame
cancel_stats = amount_of_trips.join(canceled_trips_count, how='left')

# Fill NaNs in canceled trips (users with no cancellations)
cancel_stats['canceled_trips'] = cancel_stats['canceled_trips'].fillna(0)

# Compute the canceled rate
cancel_stats['canceled_rate'] = (
    cancel_stats['canceled_trips'] / cancel_stats['total_trips']
).round(3)
cancel_stats = cancel_stats.join(has_booked['has_booked'], how='left')
```

```python
# List of all user-level DataFrames / Series created so far
feature_sources = {
    'amount_of_trips': amount_of_trips,
    'canceled_trips_count': canceled_trips_count,
    'cancel_stats': cancel_stats,
    'has_booked': has_booked,
    'browsing_counts': browsing_counts,
    'booking_counts': booking_counts,
    'browsing_behavior': browsing_behavior,
    'session_counts': session_counts,
}

# Check that each has user_id as index
for name, df in feature_sources.items():
    idx_name = df.index.name
    print(f"{name.ljust(25)} → Index: {idx_name} {'✅ OK' if idx_name == 'user_id' else '❌ MISSING!'}")
```

```python
# Total discounted bookings per user
total_discounted = valid_bookings.groupby('user_id')['any_discount_given'].sum().astype(int).rename('total_discounted_bookings')

# Create ratio without fillna (NaNs will persist for users with no total_trips or no discounts)
discount_usage = amount_of_trips.join(total_discounted, how='left')

# Calculate discount usage rate
discount_usage['discount_usage_rate'] = (
    discount_usage['total_discounted_bookings'] / discount_usage['total_trips']
).round(3)
```

```python
discount_usage
```

```python
# Cross-check against your has_booked column
discount_usage_with_book_flag = discount_usage.join(has_booked)

# Users with NaN discount usage AND no bookings
non_booking_discount_users = discount_usage_with_book_flag[
    (discount_usage_with_book_flag['discount_usage_rate'].isna()) &
    (discount_usage_with_book_flag['has_booked'] == 0)
]

print("Users with NaN discount usage and no bookings:", non_booking_discount_users.shape[0])
```

```python
non_booking_discount_users = discount_usage_with_book_flag[
    (discount_usage_with_book_flag['discount_usage_rate'].isna()) &
    (discount_usage_with_book_flag['has_booked'] == 0)
]
```

```python
# 1. Flag sessions where any discount was offered (even if not used)
valid_sessions['discount_shown'] = (
    (valid_sessions['flight_discount_amount'].fillna(0) > 0) |
    (valid_sessions['hotel_discount_amount'].fillna(0) > 0)

)

# 2. How many sessions with discount shown per user
discount_shown_sessions = (
    valid_sessions[valid_sessions['discount_shown']]
    .groupby('user_id')
    .size()
    .rename('discount_shown_sessions')
)

# 3. How many of those sessions resulted in a booking (conversion)
discount_conversion_sessions = (
    valid_sessions[valid_sessions['discount_shown'] & valid_sessions['trip_id'].notna()]
    .groupby('user_id')
    .size()
    .rename('discount_conversion_sessions')
)

# 4. Combine both counts
discount_conversion = pd.concat([
    discount_shown_sessions,
    discount_conversion_sessions
], axis=1)

# 5. Calculate the per-user conversion rate (keep NaNs for now)
discount_conversion['discount_conversion_rate'] = (
    discount_conversion['discount_conversion_sessions'] /
    discount_conversion['discount_shown_sessions']
).round(3)
```

```python
discount_conversion
```

```python
# 1. Total sessions per user (regardless of booking)
total_sessions_per_user = valid_sessions.groupby('user_id').size().rename('total_sessions')

# 2. Sessions where a discount was shown (flight or hotel)
valid_sessions['discount_shown'] = (
    (valid_sessions['flight_discount_amount'].fillna(0) > 0) |
    (valid_sessions['hotel_discount_amount'].fillna(0) > 0)

)
discount_shown_sessions = valid_sessions[valid_sessions['discount_shown']] \
    .groupby('user_id') \
    .size() \
    .rename('sessions_with_discount_shown')

# 3. Combine into one DataFrame
session_discount_exposure = pd.concat([
    total_sessions_per_user,
    discount_shown_sessions
], axis=1)

# 4. Calculate the share of sessions where a discount was shown
session_discount_exposure['discount_shown_ratio'] = (
    session_discount_exposure['sessions_with_discount_shown'] /
    session_discount_exposure['total_sessions']
).round(3)

# 5. Fill NaNs if needed (for users with no discount-shown sessions)
session_discount_exposure = session_discount_exposure.fillna(0)
```

```python
session_discount_exposure
```

```python
# Base index of all users
all_user_ids = pd.DataFrame({'user_id': sessions['user_id'].unique()}).set_index('user_id')

# Build session_summary with ONLY selected columns
session_summary = all_user_ids \
    .join(browsing_counts.rename('browsing_sessions'), how='left') \
    .join(booking_counts.rename('booking_sessions'), how='left') \
    .join(session_counts[['total_sessions', 'booking_ratio']], how='left') \
    .join(canceled_trips_count.rename(columns={'canceled_trips': 'canceled_trips'}), how='left') \
    .join(cancel_stats[['canceled_rate']], how='left') \
    .join(discount_usage[['total_discounted_bookings', 'discount_usage_rate']], how='left') \
    .join(discount_conversion[['discount_conversion_rate']], how='left') \
    .join(session_discount_exposure[['sessions_with_discount_shown', 'discount_shown_ratio']], how='left')

session_summary = session_summary.fillna(0)
# Move 'discount_conversion_rate' to the end
cols = [col for col in session_summary.columns if col != 'discount_conversion_rate'] + ['discount_conversion_rate']
session_summary = session_summary[cols]

session_summary

```

```python
session_summary['engagement_discount_score'] = (
    session_summary['discount_shown_ratio'] *
    session_summary['discount_conversion_rate'] *
    session_summary['discount_usage_rate']
) * 100
session_summary['engagement_discount_score'] = session_summary['engagement_discount_score'].round(2)
#creating a discount activity feature to outline users who are money savers
```

```python
user_features['active_months'] = (user_features['active_days'] / 30).clip(lower=1).round(2)
```

```python
user_features['sessions_per_month'] = (
    user_features['total_sessions'] / user_features['active_months']
).round(2) #feature is not based on tenure months since the data doesnt contain anything after 2023, so this would distort the actual behavioral implications - or all users churned
```

```python
def hotel_activity_tier(spm):
    if spm >= 3:
        return 'frequent planner'
    elif spm >= 1.5:
        return 'occasional planner'
    elif spm > 0.5:
        return 'rare planner'
    else:
        return 'inactive'

user_features['activity_level'] = user_features['sessions_per_month'].apply(hotel_activity_tier)
#check for hotel lovers
```

```python
#(total_hotel_spending + total_flight_spending) / tenure_months

user_features['customer_value'] = (user_features['total_spending'] / user_features['active_months']).round(2)
#customer value is money spent by time active
```

```python
def customer_type(row):
    if row['hotel_booking_count'] > 0 and row['flight_booking_count'] > 0:
        return 'Hotel+Flight'
    elif row['hotel_booking_count'] > 0:
        return 'Hotel Only'
    elif row['flight_booking_count'] > 0:
        return 'Flight Only'
    else:
        return 'Browser Only'

user_features['customer_type'] = user_features.apply(customer_type, axis=1)
#check travel preferences
```
